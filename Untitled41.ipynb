```python
def parse_query(query):
    query = query.lower()

    if "cube" in query:
        return ("cube", {"size": 10})
    elif "sphere" in query:
        return ("sphere", {"radius": 5})
    elif "cylinder" in query:
        return ("cylinder", {"radius": 5, "height": 10})

    return None, {}
```

```python
import cadquery as cq

def create_model(shape, params):
    if shape == "cube":
        return cq.Workplane("XY").box(params["size"], params["size"], params["size"])

    elif shape == "sphere":
        return cq.Workplane("XY").sphere(params["radius"])

    elif shape == "cylinder":
        return cq.Workplane("XY").circle(params["radius"]).extrude(params["height"])

    else:
        raise ValueError("Unknown shape")
```

```python
from cadquery import exporters

def export_stl(model, filename="output.stl"):
    exporters.export(model, filename)
    print(f"STL saved as {filename}")
```

```python
query = "create a rudder"

shape, params = parse_query(query)
model = create_model(shape, params)
export_stl(model, "cube.stl")
```

```python
import cadquery as cq
import math

def create_rudder(height=100, chord=40, thickness=6):
    """
    height = vertical length
    chord = front-to-back length
    thickness = thickness of rudder
    """

    # Airfoil-like 2D profile (simplified)
    points = [
        (0, 0),  # leading edge bottom
        (0.1 * chord, 0.05 * chord),
        (0.4 * chord, 0.08 * chord),
        (0.8 * chord, 0.04 * chord),
        (chord, 0),  # trailing edge
        (0.8 * chord, -0.04 * chord),
        (0.4 * chord, -0.08 * chord),
        (0.1 * chord, -0.05 * chord),
    ]

    # Create profile
    profile = cq.Workplane("XY").spline(points).close()

    # Extrude vertically в†’ rudder blade
    rudder = profile.extrude(height)

    return rudder
```

```python
from cadquery import exporters

rudder = create_rudder(height=120, chord=50, thickness=8)
exporters.export(rudder, 'rudder.stl')

print("Rudder STL generated!")
```
